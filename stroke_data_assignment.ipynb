İnme (stroke) için potansiyel risk faktörleri altında şu konular incelenmiştir:

Age (Yaş): Daha ileri yaşlarda inme riski artar.

Hypertension (Hipertansiyon): Yüksek tansiyon, inme için önemli bir risk faktörüdür.

Heart Disease (Kalp Hastalığı): Kalp hastalığı olan bireylerde inme riski daha yüksektir.

Ever Married (Evli Olmuş): Evli olma durumu, sosyal destek ve sağlık davranışları açısından dolaylı olarak etkileyebilir.

Work Type (İş Türü): Belirli meslek grupları, yaşam tarzı ve sağlık alışkanlıklarına bağlı olarak farklı riskler taşıyabilir.

Residence Type (Yaşam Alanı Türü): Kırsal veya kentsel alanlarda yaşamak, sağlık hizmetlerine erişim ve yaşam tarzı açısından riskleri etkileyebilir.

### <span style=" color:red"> How does the combination of risk factors amplify stroke likelihood?
</span>

```python
# Filter data for people with both hypertension and heart disease
filtered_df = df[(df['hypertension'] == 1) & (df['heart_disease'] == 1)]

# Calculate stroke rate
total_people = len(filtered_df)
people_with_stroke = len(filtered_df[filtered_df['stroke'] == 1])
stroke_rate = (people_with_stroke / total_people) * 100

print(f'Stroke Rate in People with Hypertension and Heart Disease: {stroke_rate:.2f}%')

# Create a DataFrame for visualization
data = {
    'Condition': ['Hypertension and Heart Disease'],
    'Stroke Rate': [stroke_rate]
}
plot_df = pd.DataFrame(data)

# Plot the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(data=plot_df, x='Condition', y='Stroke Rate', palette='Greens')
plt.xlabel('Condition')
plt.ylabel('Stroke Rate (%)')
plt.title('Stroke Rate in People with Hypertension and Heart Disease')
plt.ylim(0, 100)
plt.show()
```

### <span style=" color:red"> How are continuous variables related to categorical variables?
</span>

```python
import scipy.stats as stats

male_glucose = df_11[df_11['gender'] == 'Male']['avg_glucose_level']
female_glucose = df_11[df_11['gender'] == 'Female']['avg_glucose_level']

f_stat, p_value = stats.f_oneway(male_glucose, female_glucose)

data = [male_glucose, female_glucose]
labels = ['Male', 'Female']

fig, ax = plt.subplots()

ax.boxplot(data, labels=labels, patch_artist=True)
ax.set_title(f'Gender and Glucose Level\nANOVA p-value: {p_value:.4f}')
ax.set_ylabel('Average Glucose Level')

plt.show()
```
